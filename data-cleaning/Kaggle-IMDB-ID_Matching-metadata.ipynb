```python
import warnings
```

```python
# return (Kaggle ID, IMDB ID)
def find_ids(line, left_brackets, right_brackets, lineNum):
    if line[0] == ' ':
        warnings.warn("possibly line error at line: " + str(lineNum))
#         print(line[:20])
        return
    depth = 0
    detected_comma = 0
    even_quotes = True
    for i in range(len(line)):
        ch = line[i]
        if ch == '"':
            even_quotes = not even_quotes
        elif ch in left_brackets:
            depth += 1
        elif ch in right_brackets:
            depth -= 1 
        elif ch == ',' and depth == 0 and even_quotes:
            detected_comma += 1
        if detected_comma == 5:
            break
    if detected_comma != 5:
        raise Exception("wrong format")
    truncted = line[i:].split(',')
    if not truncted[2].startswith("tt"):
        if truncted[2] == '':
            warnings.warn("empty imdb id at line:" + str(lineNum))
#             print(truncted[1])
            return
        else:
            raise Exception("incorrect ID: i = " + str(i) + ", id: " + truncted[2])
    return truncted[1], truncted[2]
```

```python
metadata = open("movies_metadata.csv", "r")
matchings = open("matching.csv", "w")
raw_line = metadata.readline()
raw_line = metadata.readline()
left_brackets = {'{', '['}
right_brackets = {'}', ']'}
matchings.write("imdbid, kaggleid\n")
count = 2
while (raw_line):
    # manully edit the changeline errors
    ids = find_ids(raw_line, left_brackets, right_brackets, count)
    if (ids):
        kaggle_id, imdb_id = ids
        matchings.write(str(imdb_id) + "," + str(kaggle_id) + "\n")
    raw_line = metadata.readline()
    count += 1
metadata.close()
matchings.close()
print("done")
```
